```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [2,4,6,8,9]
y = [23,28,21,15,9]
plt.plot(x, y, color='red',marker='o', label='Growth Line')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(y,x, color='green',linestyle= '--',marker='o', label='Growth Line')
plt.axhline(y = max(y),color = 'red')



plt.plot(x,y)
```

```python
products = ['Books', 'Charts', 'Pens', 'Eraser', 'Paper']
count = [200, 25, 150, 35, 135]
plt.bar(products, count, color='skyblue',width = 0.4)
plt.xlabel("Products")
plt.ylabel("Quantity Sold")
plt.show()
```

```python
products = ['Books', 'Charts', 'Pens', 'Eraser', 'Paper']
count = [200, 25, 150, 35, 135]
plt.title("Pie Chart")

plt.pie(count, labels=products, autopct='%1.1f%%', shadow=True, startangle=140)
plt.show()
```

```python
import seaborn as sns
sns.get_dataset_names()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(data = iris, x='sepal_length',y = 'sepal_width',hue = 'species')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
sns.pairplot(data = iris,hue = 'species')
```

```python
sns.boxplot(data = iris,x='species',y='petal_length')
```
